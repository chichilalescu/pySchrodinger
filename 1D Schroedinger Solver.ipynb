```python
import numpy as np
import os
import sys
%matplotlib nbagg
import matplotlib
matplotlib.usetex = 1
matplotlib.use('TKAgg')

import matplotlib.pyplot as plt
from matplotlib import animation

from scipy import fftpack
from scipy import misc
from scipy import ndimage

from schrodinger import Schrodinger

############# Gaussian wave-packets
def gauss_x(x, a, x0, k0):
    """
    a gaussian wave packet of width a, centered at x0, with momentum k0
    """
    return ((a * np.sqrt(np.pi))**(-0.5) *
            np.exp(-0.5 * ((x - x0) * 1. / a)**2 + 1j * x * k0))


def gauss_k(k, a, x0, k0):
    """
    analytical fourier transform of gauss_x(x), above
    """
    return ((a / np.sqrt(np.pi))**0.5 *
            np.exp(-0.5 * (a * (k - k0))**2 - 1j * (k - k0) * x0))

def derivative(x, mu):
  n = len(x)
  xx = np.append( np.append(
                (x[1:n/2+1][::-1]), x ),
                 x[-n/2-1:-1][::-1] ) # boundaries
  nn = 2 * n
  xft = np.fft.fft(xx)
  omega = np.fft.fftfreq(nn, d = 1./n) * 2 *np.pi 
  y = np.nan_to_num(( omega * 1j) ** mu)
  y[0] = 0.
  dxdt = np.real(np.fft.ifft( y * xft ))
  dxdt = dxdt[n/2:-n/2]
  return dxdt
  
def fgm(n, H = 0.75):
  mu = 0.5 - H
  x = derivative( np.random.randn(n), mu)
  x -= x.mean()
  return x/x.std()

def fbm(n, H = 0.75):
  x = derivative(fgm(n, H), -1.)
  x -= np.mean(x)
  return x/x.std()

def ConicalPotential(x, e, zero = 10):
    v = zero**e
    return .5*((1./v)*(v - np.abs(x)**(e)))
        
def SmoothedConicalPotential(ell, x, e, zero = 10):
    return ndimage.gaussian_filter(ConicalPotential(x, e, zero), ell)     
        
def BrownianPotential(n, alpha):
    return fbm(n, H = alpha)

def SmoothedBrownianPotential(ell, n, alpha):
    return ndimage.gaussian_filter(BrownianPotential(n, alpha), ell)
```

Solving Schroedinger
-----

We supply the Schoedinger equation with initial data in the form of a minimal uncertainty wavepacket
$$\psi_0(x;x_0) = \frac{1}{\pi^{1/4}\sigma^{1/2}}\exp\left(-\frac{|x-x_0|^2}{2\sigma^2}\right)e^{i p_0x/\hbar}$$
Note that since $\hat{p}= \hbar/i \partial_x$, we have that $\hat{p}\psi_0(x) = (p_0+ i\hbar x/\sigma^2 )\psi_0(x)$. As "$\hbar\to0$" this corresponds to fixing an initial moment for the "classical particle".


Roughly speaking, this solver uses a split physical-fourier space method.  That is first it solves 
$
i \hbar \partial_t \psi = V\psi 
$
yielding 
$
\psi(x, t+\Delta t) = \psi(x,t)e^{-iV(x) \Delta t/\hbar}.
$

Then, on the fourier side it solve 
$
i \hbar \partial_t \hat{\psi} = \frac{\hbar^2k^2}{2m}\hat{\psi} 
$
yeilding 
$
\hat{\psi}(k, t+\Delta t) = \hat{\psi}(k,t)e^{-i\hbar^2k^2 \Delta t/2m}.
$

Here are more details on algorithm that it uses

1) Discretize wavefunction to a grid $\psi_n(t)= \psi(x_n,t)$, $V_n= V(x_n)$ and $\hat{\psi}_m= \hat{\psi}(k_n,t)$.

2)  Progress the system by step $\Delta t$.  This involves first computing a "half step" in $x$:
$
x: \psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

3) The calculate $\hat{\psi}_m$ from $\psi_n$ via FFT

4)  Computing a "full step" in $k$:
$
k: \hat{\psi}_m \leftarrow \hat{\psi}_m e^{-i\hbar^2k^2 \Delta t/2m}
$

5) Calculate $\psi_n$ from $\hat{\psi}_m$ via inverse FFT

6) Compute second "half-step" in $x$
$
x: \psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

7) Repeat



```python
# specify constants
hbar = .5 # planck's constant
hbar = np.float(hbar)
m = 1      # particle mass

# specify time steps and duration
dt = hbar/64
N_steps = 100
t_max = 120
frames = int(t_max / float(N_steps * dt))

# specify range in x coordinate
dx = hbar/64
xmin = -2.**9
xmax =  2.**9
x = np.arange(xmin, xmax, dx)
print x.shape
#print int(np.log2(x.shape)*x.shape[0])

d = hbar
ell = hbar**(0.5)
printableell = int(10*ell)/10.
roughness = 1.3
print d/np.sqrt(2) #Analytic Variance for initial gaussian wavepacket



V_x    = SmoothedConicalPotential(ell, x, roughness) 
BV_x   = BrownianPotential(x.shape[0], roughness)
#V_x    = np.zeros(x.shape[0]) 
#V_x    = ConicalPotential( x, roughness)
#V_x    
#V_x    = SmoothedBrownianPotential(ell, x, roughness)

x0 = 0 
p0 = 0 
k0 = p0 / hbar
psi_x0 = gauss_x(x, d, x0, k0)

initialdispersion = np.sqrt(np.sum((np.abs(psi_x0)**2)*x**2*dx) - np.sum((np.abs(psi_x0)**2)*x*dx)**2)
initialdispersion = int(1000*initialdispersion)/1000.



# define the Schrodinger object which performs the calculations
S = Schrodinger(x = x,
                psi_x0 = psi_x0,
                V_x = V_x,
                hbar = hbar,
                m = m,
                k0 = -28)

Sparabola = Schrodinger(x = x,
                psi_x0 = psi_x0,
                V_x = SmoothedConicalPotential(0, x, 2) ,
                hbar = hbar,
                m = m,
                k0 = -28)

S0 = Schrodinger(x = x,
                psi_x0 = psi_x0,
                V_x = np.zeros(x.shape[0]),
                hbar = hbar,
                m = m,
                k0 = -28)


print S.dispersion()
print initialdispersion

ax = plt.figure().add_subplot(111)
ax.plot(x, psi_x0,  label = '$\hbar = {0},\  \\Delta(0) = {1},\ \\sigma = {2}$'.format(hbar, initialdispersion,d))
ax.plot(x, V_x,  label = '$V(x) \\sim -|x|^\\alpha,\  \\alpha = {0},\ \ell = {1}$'.format(roughness, printableell))
ax.plot(x, SmoothedConicalPotential(0, x, 2),  label = '$V(x) \\sim -|x|^2$')
#ax.plot(x, BV_x,  label = 'fractional Brownian Potential')
ax.legend(loc = "best")
ax.set_xlim(-20, 20)
ax.set_ylim(0, 2)
ax.set_xlabel('$x$', fontsize = 18)
ax.set_ylabel('$\\psi_0(x)$', fontsize = 18)
plt.gcf().savefig('initialSetuphbarp5.pdf', format = 'pdf')
```

```python
# unanimated dispersion test

finaltime = int(100/hbar) #60 corresponds to 600 for h=1.  
                  #Multply by 1/factor where factor is the fraction of hbar you choose next.  
                  #i.e. for h = .5, your time is half so you multiply by 2, i.e. finaltime = 120.

dvals = np.zeros((finaltime, 3), np.float)
psi = np.zeros((finaltime, 3, S0.psi_x.shape[0]), S0.psi_x.dtype)



for step in range(finaltime):
    S.time_step(dt, N_steps)
    S0.time_step(dt, N_steps)
    Sparabola.time_step(dt, N_steps)
    dvals[step, 0] = S.dispersion()
    dvals[step, 1] = S0.dispersion()
    dvals[step, 2] = Sparabola.dispersion()
    psi[step, 0] = S.psi_x
    psi[step, 1] = S0.psi_x
    psi[step, 2] = Sparabola.psi_x
    print step
```

```python
psi1 = psi[:, 0]
psi2 = psi[:, 1]
psi3 = psi[:, 2]



dispersion1 = np.sqrt(np.sum((np.abs(psi1)**2)*x**2*dx, axis = 1) -
                    np.sum((np.abs(psi1)**2)*x*dx, axis = 1)**2)
dispersion2 = np.sqrt(np.sum((np.abs(psi2)**2)*x**2*dx, axis = 1) -
                    np.sum((np.abs(psi2)**2)*x*dx, axis = 1)**2)
dispersion3 = np.sqrt(np.sum((np.abs(psi3)**2)*x**2*dx, axis = 1) -
                    np.sum((np.abs(psi3)**2)*x*dx, axis = 1)**2)



observationtime = 20

ax = plt.figure().add_subplot(111)
#ax.plot(S.x, np.abs(psi[0, 0])**2, color = 'black')
ax.plot(S.x, np.abs(psi1[observationtime])**2, color = 'blue', dashes = (1, 1), label = '$V(x) \\sim -|x|^\\alpha$')
ax.plot(S.x, np.abs(psi2[observationtime])**2, color = 'red', dashes = (1, 1), label = '$V(x) = 0$')
ax.plot(S.x, np.abs(psi3[observationtime])**2, color = 'green', dashes = (1, 1), label = '$V(x) \\sim -|x|^2$')
ax.legend(loc = "best")
ax.set_xlim(-100, 100)
ax.set_xlabel('$x$', fontsize = 18)
ax.set_ylabel('$\\psi_0(x)$', fontsize = 18)
plt.gcf().savefig('observationhp5.pdf', format = 'pdf')
              
              


tvals = np.array(range(finaltime)).astype(np.float)*dt*N_steps
tfine = np.arange(0,finaltime, .001)
hagedorndispersion = np.sqrt(.5*tfine**2*d**2*np.cosh(tfine/100.))

ax = plt.figure().add_subplot(111)
#ax.plot(S.x, np.abs(psi[0, 0])**2, color = 'black')
ax.plot(tvals, dispersion1, color = 'blue', label = '$V(x) \\sim -|x|^\\alpha$')
ax.plot(tvals, dispersion2, color = 'green',  label = '$V(x) = 0$')
ax.plot(tvals, dispersion3, color = 'red', label = '$V(x) \\sim -|x|^2$')
ax.plot(tfine, hagedorndispersion,  color = 'cyan', label = 'Hagedorn Asymp.')
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.legend(loc = "best")
ax.set_xlabel('time $t$', fontsize = 18)
ax.set_ylabel('$\\Delta(t)$', fontsize = 18)
plt.gcf().savefig('dispersionhp5.pdf', format = 'pdf')
```

```python
fig = plt.figure()

# plotting limits
xlim = (x[0], x[-1])
klim = (-5, 5)

# top axes show the x-space data
ymin = abs(S.psi_x).min()
ymax = max(abs(S.psi_x).max(),V_x.max())
ax1 = fig.add_subplot(211, xlim = xlim,
                      ylim = (ymin - 0.2 * (ymax - ymin),
                            ymax + 0.2 * (ymax - ymin)))
psi_x_line, = ax1.plot([], [], c='r', label=r'$|\psi(x)|$')
V_x_line, = ax1.plot([], [], c='k', label=r'$V(x)$')
center_line = ax1.axvline(0, c='k', ls=':', label=r"$x_0 + v_0t$")

title = ax1.set_title("")
ax1.legend(prop=dict(size=12))
ax1.set_xlabel('$x$')
ax1.set_ylabel(r'$|\psi(x)|$')

# bottom axes show the k-space data
ymin = abs(S.psi_k).min()
ymax = abs(S.psi_k).max()
ax2 = fig.add_subplot(212, xlim=klim,
                      ylim=(ymin - 0.2 * (ymax - ymin),
                            ymax + 0.2 * (ymax - ymin)))
psi_k_line, = ax2.plot([], [], c='r', label=r'$|\psi(k)|$')

p0_line1 = ax2.axvline(-p0 / hbar, c='k', ls=':', label=r'$\pm p_0$')
p0_line2 = ax2.axvline(p0 / hbar, c='k', ls=':')
#mV_line = ax2.axvline(np.sqrt(2 * V0) / hbar, c='k', ls='--',
#                      label=r'$\sqrt{2mV_0}$')
ax2.legend(prop=dict(size=12))
ax2.set_xlabel('$k$')
ax2.set_ylabel(r'$|\psi(k)|$')

V_x_line.set_data(S.x, S.V_x)



######################################################################
# Functions to Animate the plot
def init():
    psi_x_line.set_data([], [])
    V_x_line.set_data([], [])
    center_line.set_data([], [])

    psi_k_line.set_data([], [])
    title.set_text("")
    return (psi_x_line, V_x_line, center_line, psi_k_line, title)


def animate(i):
    S.time_step(dt, N_steps)
    psi_x_line.set_data(S.x, 4 * abs(S.psi_x))
    V_x_line.set_data(S.x, S.V_x)
    center_line.set_data(2 * [x0 + S.t * p0 / m], [0, 1])

    psi_k_line.set_data(S.k, abs(S.psi_k))
    title.set_text("t = %.2f" % S.t)
    return (psi_x_line, V_x_line, center_line, psi_k_line, title)

# call the animator.
# blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=frames, interval=30, blit=True)


# uncomment the following line to save the video in mp4 format.  This
# requires either mencoder or ffmpeg to be installed on your system
#anim.save('schrodinger_barrier.mp4', fps=15,
#          extra_args=['-vcodec', 'libx264'])

plt.show()
```
