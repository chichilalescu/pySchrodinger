```python
import numpy as np
import os
import sys
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import animation

import subprocess
import pickle
import gzip
import os

matplotlib.usetex = 1
matplotlib.use('TKAgg')

from scipy import fftpack
from scipy import misc
from scipy import ndimage

from schrodinger import Schrodinger

######################### Gaussian wave-packet #################################

def beta(t, a, h, m):
    return 1 + 1j*(h*t)/(m*a**2)

def E(k0, m):
    return k0**2/(2*m)

def gaussian_wavepacket(t, x, a, h, m, x0, k0):
    beta =  1 + 1j*(h*t)/(m*a**2)
    E = k0**2/(2*m)
    return ((a*beta*np.sqrt(np.pi))**(-0.5)*
               np.exp(1j*(k0*x - E*t)/h) *
                    np.exp(-0.5*((x - x0 - k0*t/m)*1.)**2/(a**2**beta)))

######################### Fractional Brownian Potential ######################### 

def derivative(x, mu):
    n = len(x)
    xx = np.append( np.append(
                (x[1:n/2+1][::-1]), x ),
                 x[-n/2-1:-1][::-1] ) # boundaries
    nn = 2 * n
    xft = np.fft.fft(xx)
    omega = np.fft.fftfreq(nn, d = 1./n) * 2 *np.pi 
    y = np.nan_to_num(( omega * 1j) ** mu)
    y[0] = 0.
    dxdt = np.real(np.fft.ifft( y * xft ))
    dxdt = dxdt[n/2:-n/2]
    return dxdt
  
def fbm(n, H = 0.75, rseed = None):
    mu = 0.5 - H
    np.random.seed(rseed)
    x = derivative( np.random.randn(n), mu)
    x -= x.mean()
    y = derivative(x/x.std(), -1.)
    y -= np.mean(y)
    return y/y.std()

def SmoothedBrownianPotential(ell, n, alpha, rseed = None):
    bla = fbm(n, H = alpha, rseed = rseed)
    return ndimage.gaussian_filter(bla, ell)

########################## Cusp Potential #######################################

def LocalizedSmoothConicalPotential(ell, x, e, zero = 1):
    v = zero**e
    ConicalPotential = .5*((1./v)*(v - np.abs(x)**(e)))
    bumpfunction = np.piecewise(x, [abs(x) < zero, abs(x) >= zero], [1, 0])
    return  ndimage.gaussian_filter(ConicalPotential*bumpfunction, ell)  
        
```

Solving Schroedinger
-----

We supply the Schoedinger equation with initial data in the form of a minimal uncertainty wavepacket
$$\psi_0(x;x_0) = \frac{1}{(2\pi)^{1/4}\sigma^{1/2}}\exp\left(-\frac{|x-x_0|^2}{4\sigma^2}\right)e^{i p_0x/\hbar}$$
Note that since $\hat{p}= \hbar/i \partial_x$, we have that $\hat{p}\psi_0(x) = (p_0+ i\hbar x/\sigma^2 )\psi_0(x)$. As "$\hbar\to0$" this corresponds to fixing an initial moment for the "classical particle".


Roughly speaking, this solver uses a split physical-fourier space method.  That is first it solves 
$
i \hbar \partial_t \psi = V\psi 
$
yielding 
$
\psi(x, t+\Delta t) = \psi(x,t)e^{-iV(x) \Delta t/\hbar}.
$

Then, on the fourier side it solve 
$
i \hbar \partial_t \hat{\psi} = \frac{\hbar^2k^2}{2m}\hat{\psi} 
$
yeilding 
$
\hat{\psi}(k, t+\Delta t) = \hat{\psi}(k,t)e^{-i\hbar^2k^2 \Delta t/2m}.
$

Here are more details on algorithm that it uses

1) Discretize wavefunction to a grid $\psi_n(t)= \psi(x_n,t)$, $V_n= V(x_n)$ and $\hat{\psi}_m= \hat{\psi}(k_n,t)$.

2)  Progress the system by step $\Delta t$.  This involves first computing a "half step" in $x$:
$
x: \psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

3) The calculate $\hat{\psi}_m$ from $\psi_n$ via FFT

4)  Computing a "full step" in $k$:
$
k: \hat{\psi}_m \leftarrow \hat{\psi}_m e^{-i\hbar^2k^2 \Delta t/2m}
$

5) Calculate $\psi_n$ from $\hat{\psi}_m$ via inverse FFT

6) Compute second "half-step" in $x$
$
\psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

7) Repeat



```python
def Solve_Schroedinger(hbar = 1, #Planck's constant
                       t_max = 10,
                       m = 1, #particle mass
                       x0 = 0, #center of initial wavefunction
                       p0 = 0, #momentum of initial wavefunction
                       roughness = 1.3, #exponent of conical potential
                       xmin = -2.**6*np.pi, 
                       xmax =  2.**6*np.pi,
                       N = 16):
    
    hbar = np.float(hbar)
    sigma = 1*hbar**(0.5) 
    ell = 2**7*hbar**2
    zeroOfPotential = 15
    
    TT = int(N/hbar)
    dt = TT**(-1)
    N_steps = int(t_max/dt)
    dx = np.pi*hbar/(N) 
    xmax_new = int(xmax/dx)*dx
    xmin_new = int(xmin/dx)*dx
    x = np.arange(xmin_new, xmax_new, dx)
    
    psi_x0 = gaussian_wavepacket(0, x, sigma, hbar, 1, x0, 0)
    

    S = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = LocalizedSmoothConicalPotential(ell, x, roughness, zeroOfPotential),
                    hbar = hbar,
                    m = m)
    
    S0 = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = np.zeros(x.shape[0]),
                    hbar = hbar,
                    m = m)

    S1 = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = LocalizedSmoothConicalPotential(0, x, 2, zeroOfPotential),
                    hbar = hbar,
                    m = m)
    

    psi = np.zeros((int(t_max)+1, 3, S0.psi_x.shape[0]), S0.psi_x.dtype)
    steps_per_unit_time = int(1./dt)

    psi[0, 0] = psi_x0;
    psi[0, 1] = psi_x0;
    psi[0, 2] = psi_x0;
    
    for step in range(int(t_max)):
        S.time_step(dt, steps_per_unit_time)
        S0.time_step(dt, steps_per_unit_time)
        S1.time_step(dt, steps_per_unit_time)
        psi[step + 1, 0] = S.psi_x
        psi[step + 1, 1] = S0.psi_x
        psi[step + 1, 2] = S1.psi_x
        print step + 1
    S.psi_x_full = psi[:, 0]
    S0.psi_x_full = psi[:, 1]
    S1.psi_x_full = psi[:, 2]
    
    print S.k
    print np.shape(S.k)

    for solver in [S, S0, S1]:
        solver.dispersion_vs_t = \
            np.sqrt(np.sum((np.abs(solver.psi_x_full)**2)*x**2*dx, axis = 1) -
                    np.sum((np.abs(solver.psi_x_full)**2)*x*dx, axis = 1)**2)
    return  S, S0, S1
```

```python
hbarUsed = np.array([2**(0), 2**(-1), 2**(-2), 2**(-3), 2**(-4), 2**(-5), 2**(-6), 2**(-7), 2**(-8), 2**(-9)])
#hbarUsed = np.array([2**(-10)])

class GetData:
    def __init__(self, T = 20, N = 2**6):
        self.T = T
        self.N = N
        self.sol = {}
        return None
    def compute(self, h):
        print('solving Schroedinger for hbar = {0}'.format(h))
        fname = 'data/sol_h{0}_T{1}_N{2}.pickle.gzip'.format(h, self.T, self.N)
        if os.path.exists(fname):
            self.sol['{0}'.format(h)] = pickle.load(gzip.open(fname, 'r'))
        else:
            self.sol['{0}'.format(h)] = Solve_Schroedinger(hbar = h,
                                                           t_max = self.T,
                                                           N = self.N)
            pickle.dump(self.sol['{0}'.format(h)],
                        gzip.open(fname, 'w'))
        return self.sol['{0}'.format(h)]
computator = GetData(T = 30, N = 2**6)
storedData = [computator.compute(h) for h in hbarUsed]
```

Wavefunction Measurement and Dispersion
---

In the case of a free Gaussian wavepacket, we can solve Schroedinger's equation analytically.  The result is the following
\begin{equation}
\psi(x,t) = \frac{1}{\pi^{1/4}\sqrt{\sigma\beta(t)}} e^{i(p_0x-E(p_0)t)/\hbar} e^{-[x-x_0-(p_0/m)t]^2/(2\sigma^2\beta(t))}
\end{equation}
where $\beta(t)= 1 + i\frac{\hbar t}{m \sigma^2}$ and $E(p_0)= p_0^2/2m$. Showing our evolution of a free wavepacket agrees with the analytic form will be a check on our code.  The analytic form for the dispersion for this free wave packet is
\begin{equation}
\Delta x(t) = \frac{\sigma}{\sqrt{2}} \sqrt{1 + \left(\frac{\hbar t}{m \sigma^2}\right)^2}
\end{equation}
On the other hand, for a particle moving in a parabolic potential, the dispersion in the semiclassical limit may be computed analytically via Hagedorn's equations.  This reads
\begin{equation}
\Delta x(t) = \frac{t \sigma}{\sqrt{2}} \cosh(c t)
\end{equation}
where $c$ is (twice) constant in front of the $x^2$ term in of the parabola.

```python
def packet_dispersion(t, a, h, m):
    return (a/np.sqrt(2))*np.sqrt(1 + (h*t/(m*a**2))**2)

def squared_wavefunction(t, x, a, h, m, p0):
    beta = 1 + 1j*(h*t)/(m*a**2)
    return (1/((np.pi)**(1/2.)*a*abs(beta))*np.exp(-(x - (p0/m)*t)**2/(a**2 * abs(beta**2))

def ExtremalCharacteristics(t, alpha):
    constant0 = ((1-alpha)**2/2.)**(1-alpha)
    exponent = 2/(1-alpha)
    return constant0*t**exponent, - constant0*t**exponent


def snapshop_wavefunction(
        tf = 20,
        data = storedData[0],
        minx = -20,
        maxx = 20,
        yheight = 20,
        prefix = 'bla'):
    h = data[0].hbar
    a = np.sqrt(h)
    
    time = np.arange(0, data[0].dispersion_vs_t.shape[0]).astype(np.float)
    #Analytical formulas
    
    ExtCharLeft, ExtCharRight = ExtremalCharacteristics(time, 0.3)
    
    hagedorndispersion = 1/np.sqrt(2)*time*a*np.cosh(time/np.sqrt(10)) #parabolic potential, valid in semiclassical limit
    lineardispersion = packet_dispersion(time, a, h, 1)

    fig = plt.figure(figsize=(10, 4))
    axw = fig.add_subplot(121)
    axd = fig.add_subplot(122)
    for observation_time in range(tf):
        axw.cla()
        axd.cla()
        axw.plot(data[0].x,
                 np.abs(data[0].psi_x_full[observation_time])**2,'b-',
                 color = 'blue')
        axw.plot(data[0].x,
                 squared_wavefunction(observation_time, data[0].x, a, h, 1, 0),'r--',
                 #label = '$V = 0 \ {\\rm analytic}$',
                 color = 'red')
        axw.plot(data[0].x,
                 np.abs(data[1].psi_x_full[observation_time])**2,'b-',
                 color = 'green')
        axw.plot(data[0].x,
                 np.abs(data[2].psi_x_full[observation_time])**2,'b-',
                 color = 'cyan')
        axw.plot(data[0].x, 
                 data[0].V_x - 0.45,
                 color = 'black')
        #matplotlib.pyplot.axvline(ExtCharLeft[observation_time], color = 'red', linestyle='dashed', linewidth=2)
        #matplotlib.pyplot.axvline(ExtCharRight[observation_time], color = 'red', linestyle='dashed', linewidth=2)
        axw.set_title("$\\hbar$ = {0}".format(h))
        axw.legend(loc = "best")
        axw.set_xlim(minx, maxx)
        axw.set_ylim(0, .1)
        axw.set_xlabel('$x$', fontsize = 18)
        axw.set_ylabel('$\\psi_t^*\\psi_t$', fontsize = 18)

        axd.plot(time[:observation_time+1],
                data[0].dispersion_vs_t[:observation_time+1],
                label = '$V\\sim -|x|^{(1+ \\alpha)}$',
                color = 'blue')
        axd.plot(time[:observation_time+1],
                data[1].dispersion_vs_t[:observation_time+1],
                label = '$V = 0$',
                color = 'green')
        axd.plot(time[:observation_time+1],
                data[2].dispersion_vs_t[:observation_time+1],
                label = '$V\\sim -|x|^2$',
                color = 'cyan')
        axd.plot(time, lineardispersion,'--',  color = 'red')
        axd.set_title("$\\hbar$ = {0}".format(h))
        axd.set_xlim(0, tf)
        axd.set_ylim(0, yheight)
        axd.legend(loc = "best")
        axd.set_xlabel('$t$', fontsize = 18)
        axd.set_ylabel('$|\\Delta(t)|^2$', fontsize = 18)
        fig.savefig('figs/' + prefix + '_frame_{0:0>2}.png'.format(observation_time),
                    format = 'png')
    subprocess.call(['convert', 'figs/' + prefix + '_frame*.png', prefix + '.gif'])
```

#### We calculate the dispersion of the wavefuntion $\psi(t)$ in the following way: \begin{equation} \Delta_{\psi}(t) = \left(\int_{\mathbb{R}} \psi(t)^* x^2\psi(t) {\rm d} x - \left|\int_{\mathbb{R}} \psi(t)^* x \psi(t) {\rm d} x \right|^2 \right)^{1/2} \end{equation}

```python
def plotResults(T = 30, domainsize = 40, yH = 20):
    for i in range(len(storedData)):
        snapshop_wavefunction(prefix = '{0}'.format(i),
                              tf = T,
                              data = storedData[i],
                              minx = -domainsize,
                              maxx = domainsize,
                              yheight = yH)
    return None
plotResults()
```
