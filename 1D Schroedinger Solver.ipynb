```python
import numpy as np
import os
import sys
%matplotlib nbagg
import matplotlib
matplotlib.usetex = 1
matplotlib.use('TKAgg')

import matplotlib.pyplot as plt
from matplotlib import animation

from scipy import fftpack
from scipy import misc
from scipy import ndimage

from schrodinger import Schrodinger

############# Gaussian wave-packets
def gauss_x(x, a, x0, k0):
    """
    a gaussian wave packet of width a, centered at x0, with momentum k0
    """
    return ((a * np.sqrt(np.pi))**(-0.5) *
            np.exp(-0.5 * ((x - x0) * 1. / a)**2 + 1j * x * k0))


def gauss_k(k, a, x0, k0):
    """
    analytical fourier transform of gauss_x(x), above
    """
    return ((a / np.sqrt(np.pi))**0.5 *
            np.exp(-0.5 * (a * (k - k0))**2 - 1j * (k - k0) * x0))

def derivative(x, mu):
  n = len(x)
  xx = np.append( np.append(
                (x[1:n/2+1][::-1]), x ),
                 x[-n/2-1:-1][::-1] ) # boundaries
  nn = 2 * n
  xft = np.fft.fft(xx)
  omega = np.fft.fftfreq(nn, d = 1./n) * 2 *np.pi 
  y = np.nan_to_num(( omega * 1j) ** mu)
  y[0] = 0.
  dxdt = np.real(np.fft.ifft( y * xft ))
  dxdt = dxdt[n/2:-n/2]
  return dxdt
  
def fgm(n, H = 0.75):
  mu = 0.5 - H
  x = derivative( np.random.randn(n), mu)
  x -= x.mean()
  return x/x.std()

def fbm(n, H = 0.75):
  x = derivative(fgm(n, H), -1.)
  x -= np.mean(x)
  return x/x.std()

def ConicalPotential(x, e, zero = 10):
    v = zero**e
    return .5*((1./v)*(v - np.abs(x)**(e)))
        
def SmoothedConicalPotential(ell, x, e, zero = 10):
    return ndimage.gaussian_filter(ConicalPotential(x, e, zero), ell)     
        
def BrownianPotential(n, alpha):
    return fbm(n, H = alpha)

def SmoothedBrownianPotential(ell, n, alpha):
    return ndimage.gaussian_filter(BrownianPotential(n, alpha), ell)
```

Solving Schroedinger
-----

We supply the Schoedinger equation with initial data in the form of a minimal uncertainty wavepacket
$$\psi_0(x;x_0) = \frac{1}{\pi^{1/4}\sigma^{1/2}}\exp\left(-\frac{|x-x_0|^2}{2\sigma^2}\right)e^{i p_0x/\hbar}$$
Note that since $\hat{p}= \hbar/i \partial_x$, we have that $\hat{p}\psi_0(x) = (p_0+ i\hbar x/\sigma^2 )\psi_0(x)$. As "$\hbar\to0$" this corresponds to fixing an initial moment for the "classical particle".


Roughly speaking, this solver uses a split physical-fourier space method.  That is first it solves 
$
i \hbar \partial_t \psi = V\psi 
$
yielding 
$
\psi(x, t+\Delta t) = \psi(x,t)e^{-iV(x) \Delta t/\hbar}.
$

Then, on the fourier side it solve 
$
i \hbar \partial_t \hat{\psi} = \frac{\hbar^2k^2}{2m}\hat{\psi} 
$
yeilding 
$
\hat{\psi}(k, t+\Delta t) = \hat{\psi}(k,t)e^{-i\hbar^2k^2 \Delta t/2m}.
$

Here are more details on algorithm that it uses

1) Discretize wavefunction to a grid $\psi_n(t)= \psi(x_n,t)$, $V_n= V(x_n)$ and $\hat{\psi}_m= \hat{\psi}(k_n,t)$.

2)  Progress the system by step $\Delta t$.  This involves first computing a "half step" in $x$:
$
x: \psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

3) The calculate $\hat{\psi}_m$ from $\psi_n$ via FFT

4)  Computing a "full step" in $k$:
$
k: \hat{\psi}_m \leftarrow \hat{\psi}_m e^{-i\hbar^2k^2 \Delta t/2m}
$

5) Calculate $\psi_n$ from $\hat{\psi}_m$ via inverse FFT

6) Compute second "half-step" in $x$
$
x: \psi_n \leftarrow \psi_ne^{-iV(x) (\Delta t/2)/\hbar}
$

7) Repeat



```python
def Solve_Schroedinger(hbar = 1, #Planck's constant
                       t_max = 10,
                       m = 1, #particle mass
                       x0 = 0, #center of initial wavefunction
                       p0 = 0, #momentum of initial wavefunction
                       roughness = 1.3, #exponent of conical potential
                       xmin = -2.**9, 
                       xmax =  2.**9):
    
    hbar = np.float(hbar)
    sigma = hbar**(0.5) 
    ell = hbar**(0.5)
    k0 = p0/hbar
    
    # 64 is quite slow
    dt = hbar/16 
    N_steps = int(t_max/dt)
    dx = hbar/16
    x = np.arange(xmin, xmax, dx)
    
    psi_x0 = gauss_x(x, sigma, x0, k0)

    S = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = SmoothedConicalPotential(ell, x, roughness) ,
                    hbar = hbar,
                    m = m,
                    k0 = -28)
    
    S0 = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = np.zeros(x.shape[0]),
                    hbar = hbar,
                    m = m,
                    k0 = -28)

    S1 = Schrodinger(x = x,
                    psi_x0 = psi_x0,
                    V_x = ConicalPotential(x, 2) ,
                    hbar = hbar,
                    m = m,
                    k0 = -28)


    psi = np.zeros((int(t_max), 3, S0.psi_x.shape[0]), S0.psi_x.dtype)
    steps_per_unit_time = int(1./dt)
    for step in range(int(t_max)):
        S.time_step(dt, steps_per_unit_time)
        S0.time_step(dt, steps_per_unit_time)
        S1.time_step(dt, steps_per_unit_time)
        psi[step, 0] = S.psi_x
        psi[step, 1] = S0.psi_x
        psi[step, 2] = S1.psi_x
        print step
    S.psi_x_full = psi[:, 0]
    S0.psi_x_full = psi[:, 1]
    S1.psi_x_full = psi[:, 2]

    for solver in [S, S0, S1]:
        solver.dispersion_vs_t = \
            np.sqrt(np.sum((np.abs(solver.psi_x_full)**2)*x**2*dx, axis = 1) -
                    np.sum((np.abs(solver.psi_x_full)**2)*x*dx, axis = 1)**2)
    return  S, S0, S1
```

```python
T = 30
h1p0data = Solve_Schroedinger(hbar = 1,  t_max = T)
h0p5data = Solve_Schroedinger(hbar = .5, t_max = T)
h0p1data = Solve_Schroedinger(hbar = .1, t_max = T)
```

```python
def evolved_wavepacket(t, x, a, x0, h, m):
    return (a * np.sqrt(np.pi))**(-0.5)*(a**2/(a**2 + 1j*h*t/m))**(3/2.)*np.exp(-(((x - x0) * 1.)**2 / 2*(a**2 + 1j*h*t/m)))


def snapshop_wavefunction(observation_time = 10, data = h1p0data, minx = -20, maxx = 20):
    ax = plt.figure().add_subplot(111)
    ax.plot(data[0].x,
            np.abs(data[0].psi_x_full[observation_time])**2,
            label = '$V\\sim x^\\alpha$',
            color = 'blue')
    ax.plot(data[0].x,
            np.abs(data[1].psi_x_full[observation_time])**2,
            label = '$V=0$',
            color = 'green')
    ax.plot(data[0].x,
            np.abs(data[2].psi_x_full[observation_time])**2,
            label = '$V\\sim x^2$',
            color = 'red')
#    ax.plot(data[0].x,
#            np.abs(evolved_wavepacket(observation_time, data[0].x, np.sqrt(data[1].hbar), 0, data[1].hbar, 1))**2,
#            label = '$V=0$',
#            color = 'orange')
    ax.plot(data[0].x, 
            data[0].V_x,
            color = 'black')
#    ax.plot(data[0].x, 
#            data[2].V_x,
#            color = 'black')
    ax.legend(loc = "best")
    ax.set_xlim(minx, maxx)
    ax.set_ylim(0, 1)
    ax.set_xlabel('$x$', fontsize = 18)
    ax.set_ylabel('$\\psi_0(x)$', fontsize = 18)
    plt.gcf().savefig('snapshot_psi{0}.pdf'.format(data), format = 'pdf')

    
def dispersion(tf = 10, data = h1p0data):
    a = np.sqrt(data[0].hbar)
    time = np.arange(0, data[0].dispersion_vs_t.shape[0]).astype(np.float)
    #Analytical formulas
    hagedorndispersion = np.sqrt(.5*time**2*a**2*np.cosh(time/100.)) #parabolic potential, valid in semiclassical limit
    lineardispersion = np.sqrt((a**2 + (data[0].hbar*time)**2)/a) #no potential gaussian initial wavepacket
    
    ax = plt.figure().add_subplot(111)
    print time.shape, data[0].dispersion_vs_t.shape
    ax.plot(time,
            data[0].dispersion_vs_t,
            label = '$V\\sim x^\\alpha$',
            color = 'blue')
    ax.plot(time,
            data[1].dispersion_vs_t,
            label = '$V=0$',
            color = 'green')
    ax.plot(time,
            data[2].dispersion_vs_t,
            label = '$V\\sim x^2$',
            color = 'red')
    ax.plot(time, hagedorndispersion,  color = 'cyan', label = 'Hagedorn Asymp.')
    ax.plot(time, lineardispersion,  color = 'black', label = 'Free Wavepacket')
    ax.set_xlim(0, tf)
    ax.set_ylim(0, 30)
    ax.legend(loc = "best")
    ax.set_xlabel('time $t$', fontsize = 18)
    ax.set_ylabel('$\\Delta(t)$', fontsize = 18)
    plt.gcf().savefig('dispersion{0}.pdf'.format(data), format = 'pdf')  
```

#### We calculate the dispersion of the wavefuntion $\psi(t)$ in the following way: \begin{equation} \Delta_{\psi}(t) = \left(\int_{\mathbb{R}} \psi(t)^* x^2\psi(t) {\rm d} x - \left|\int_{\mathbb{R}} \psi(t)^* x \psi(t) {\rm d} x \right|^2 \right)^{1/2} \end{equation}

```python
T = 20
dataset = h1p0data
snapshop_wavefunction(observation_time = T, data = dataset,  minx = -20, maxx = 20)
dispersion(tf = T, data = dataset)
```
